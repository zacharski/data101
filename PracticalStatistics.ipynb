We've created a Pandas dataframe from a file before, but I haven't really explained it.  The Pandas function that loads a file is:

    pd.read_csv(filename)
    
So for example:

    tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
    

will read the file tips.csv and create a pandas dataframe we call `tips`.

As the name suggests `read_csv` reads a csv file. **csv** stands for comma separated values. If you click on [this link to the tips.csv file](https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv), you will see that the file looks like:

CSV files are a very common format and every spreadsheet program from Microsoft Excel to Google Sheets can read and write them. Plus, they are one of the standard file formats for data science. 

The pandas function `read_csv` can read csv files from the web, as we just did, and it can also read files directly from your computer. # Some plot examples

Now that we have a data set we can review our knowledge of making plots and charts.


## Scatter Plot
Let's do a Scatter Plot of the total bill :

```python

import seaborn as sns
sns.set_style("whitegrid")
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Tips in Relation To The Total Bill')
plt.ylabel('Tip ($)')
plt.xlabel('Total Bill ($)')
plt.show()
```

## Bar Plot
Let's still use total bill but plot it as a bar plot using the default binning:


```python
import matplotlib.pyplot as plt
_ = plt.hist(tips['total_bill'])
_ = plt.title('Histogram of Total Bill')
_ = plt.xlabel('total bill')
_ = plt.ylabel('number of customers')
plt.show()
```

```python
ax = sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Mean Total Bill')
plt.show()
```

This shows the average bill for each of the days. So, for example, the average bill on sunday was around $21.00. The whiskers indicate uncertainty. Basically, if this was just a sample of all customers, then we are 95% certain that the average bill will be within the whisker.  

#### The sum of all the bills
Suppose we are interested in the total of the bills for each day (instead of the average or mean). For that we can use the `estimator=sum` argument (The `ci` argument gets rid of the whiskers): 

```python

ax = sns.barplot(x="day", y="total_bill", data=tips, estimator=sum, palette="Blues_d",  ci=None)
plt.title('Total Sales Per Day')
plt.show()
```

  So we have the most sales on Saturday.

### Slightly fancier bar plots
Do women or men spend more?

```python
ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title('Mean Total Bill By Day and Sex')
plt.margins(0.2)
plt.show()
```

So the average bill of a man is greater than the average bill of a woman. Suppose we are interested in whether there are more women or men on the various days. For that we switch to a countplot:


```python
sns.set(style="darkgrid")
plt.title('Number of Customers Per Day')
ax = sns.countplot(x="day",  hue="sex", data=tips)
plt.show()
```

Wow. That is pretty interesting. There are more women customers then men customers on Thursdays. And it seems like Friday is our slowest day.

There are more examples of bar plots on the [Seaborn site](https://seaborn.pydata.org/generated/seaborn.barplot.html)

## Bee Swarm



```python
sns.swarmplot(x='sex', y='total_bill', data= tips)
plt.title('Total Bill By Sex')
plt.xlabel('sex')
plt.ylabel('total bill')
plt.show()
```

## Violin

```python
sns.violinplot(x='sex', y='total_bill', data= tips, palette='pastel')
plt.title('Total Bill By Sex')
plt.xlabel('sex')
plt.ylabel('total bill')
plt.show()
```

## ECDF


```python
import numpy as np
x = np.sort(tips['total_bill'])
y = np.arange(1, len(x)+ 1) / len(x)
plt.plot(x, y, marker='.', linestyle='none')
plt.title('ECDF Plot of Total Bill')
plt.xlabel('total bill')
plt.ylabel('ECDF')
plt.margins(0.02) # keeps data off plot edges
plt.show()
```

## Box Plot

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
plt.title('Total Bill by Sex')
plt.xlabel('sex')
plt.ylabel('Total Bill')
plt.show()
```
